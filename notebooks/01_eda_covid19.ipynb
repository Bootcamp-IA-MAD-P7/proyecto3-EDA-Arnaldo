```python
url = "https://api.covidtracking.com/v1/us/daily.csv"
response = requests.get(url)

with open("../data/us_daily.csv", "wb") as f:
    f.write(response.content)

df = pd.read_csv("../data/us_daily.csv")
print(df.shape)
df.head()
```

```python
# Conteo de valores nulos por columna, ordenado
df.isnull().sum().sort_values(ascending=False)
```

```python
# Convertir la columna 'date' a formato fecha real
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')

# Ordenar por fecha (de más antigua a más reciente)
df = df.sort_values('date').reset_index(drop=True)

# Ver el rango de fechas que cubre el dataset
print(df['date'].min(), "→", df['date'].max())
```

```python
# 1. Eliminar columnas sin valor analítico
df = df.drop(columns=['recovered', 'hash', 'dateChecked', 'lastModified'])

# 2. Convertir la fecha
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')

# 3. Ordenar cronológicamente (la API suele venir en orden descendente)
df = df.sort_values('date').reset_index(drop=True)

# 4. Verificar el rango temporal
print("Rango de fechas:", df['date'].min(), "→", df['date'].max())

# 5. Revisar cómo quedaron los nulos
df.isnull().sum().sort_values(ascending=False)
```

```python
# Verificar la fecha (si ya se convirtió, esto no hace nada raro)
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')

df = df.sort_values('date').reset_index(drop=True)

print("Rango de fechas:", df['date'].min(), "→", df['date'].max())

df.isnull().sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(df['date'], df['positive'], color='#1D9E75')
ax.set_title('Casos positivos acumulados de Covid-19 en EE.UU.')
ax.set_xlabel('Fecha')
ax.set_ylabel('Casos acumulados')
plt.tight_layout()
plt.show()
```

```python
df['positiveIncrease_7d'] = df['positiveIncrease'].rolling(7).mean()

fig, ax = plt.subplots(figsize=(14, 5))
ax.bar(df['date'], df['positiveIncrease'], alpha=0.3, label='Casos diarios', color='#85B7EB')
ax.plot(df['date'], df['positiveIncrease_7d'], color='#0C447C', linewidth=2, label='Media móvil 7 días')
ax.set_title('Casos nuevos diarios de Covid-19 en EE.UU.')
ax.set_xlabel('Fecha')
ax.set_ylabel('Casos nuevos')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
df['deathIncrease_7d'] = df['deathIncrease'].rolling(7).mean()

fig, ax = plt.subplots(figsize=(14, 5))
ax.bar(df['date'], df['deathIncrease'], alpha=0.3, label='Muertes diarias', color='#F0997B')
ax.plot(df['date'], df['deathIncrease_7d'], color='#993C1D', linewidth=2, label='Media móvil 7 días')
ax.set_title('Muertes diarias por Covid-19 en EE.UU.')
ax.set_xlabel('Fecha')
ax.set_ylabel('Muertes nuevas')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
df['cfr'] = df['death'] / df['positive'] * 100

fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(df['date'], df['cfr'], color='#993556')
ax.set_title('Tasa de mortalidad acumulada (CFR) - Covid-19 EE.UU.')
ax.set_xlabel('Fecha')
ax.set_ylabel('CFR (%)')
plt.tight_layout()
plt.show()
```

```python
df_clean_cfr = df[df['date'] >= '2020-04-01']

fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(df_clean_cfr['date'], df_clean_cfr['cfr'], color='#993556')
ax.set_title('Tasa de mortalidad acumulada (CFR) - Covid-19 EE.UU. (desde abril 2020)')
ax.set_xlabel('Fecha')
ax.set_ylabel('CFR (%)')
ax.axhline(y=df_clean_cfr['cfr'].iloc[-1], color='gray', linestyle='--', alpha=0.5, label=f"CFR final: {df_clean_cfr['cfr'].iloc[-1]:.2f}%")
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Tasa de positividad de tests (diaria)
df['positivity_rate'] = (df['positiveIncrease'] / df['totalTestResultsIncrease']) * 100
df['positivity_rate_7d'] = df['positivity_rate'].rolling(7).mean()

fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(df['date'], df['positivity_rate_7d'], color='#534AB7')
ax.set_title('Tasa de positividad de tests (media móvil 7 días) - EE.UU.')
ax.set_xlabel('Fecha')
ax.set_ylabel('Positividad (%)')
ax.set_ylim(0, 30)
plt.tight_layout()
plt.show()
```

```python
# Pacientes hospitalizados actualmente
fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(df['date'], df['hospitalizedCurrently'], color='#D85A30')
ax.set_title('Pacientes hospitalizados actualmente por Covid-19 - EE.UU.')
ax.set_xlabel('Fecha')
ax.set_ylabel('Pacientes hospitalizados')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

cols_corr = ['positiveIncrease', 'deathIncrease', 'hospitalizedCurrently', 
              'inIcuCurrently', 'onVentilatorCurrently', 'positivity_rate', 'totalTestResultsIncrease']

corr_matrix = df[cols_corr].corr()

fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f', ax=ax)
ax.set_title('Matriz de correlación entre variables clave - Covid-19 EE.UU.')
plt.tight_layout()
plt.show()
```
